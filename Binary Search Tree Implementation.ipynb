```python
class Node:
    
    def __init__(self,key):
        self.key=key
        self.lchild= None
        self.rchild = None
        
    def insert(self,data):
        if self.key is None:
            self.key=data
        if self.key == data:
            return
        if self.key>data:
            if self.lchild:
                self.lchild.insert(data)
            else:
                self.lchild=Node(data)
        else:
            if self.rchild:
                self.rchild.insert(data)
            else:
                self.rchild=Node(data)
                
    def Search(self,data):
        if self.key is None:
            print("Node is Empty!")
            return
        if self.key==data:
            print("Node is Found!")
            return
        if self.key<data:
            if self.lchild:
                self.lchild.Search(data)
            else:
                print("Node is not Found!")
        else:
            if self.rchild:
                self.rchild.Search(data)
            else:
                print("Node is not Found!")
                
    def deletion(self,data):
        if self.key is None:
            print("Tree is Empty, nothing to delete")
            return
        if self.key>data:
            if self.lchild:
                self.lchild=self.lchild.deletion(data)
            else:
                print("Node is not found in the Tree!")
        elif self.key<data:
            if self.rchild:
                self.rchild=self.rchild.deletion(data)
            else:
                print("Node is not found in the Tree")
        else:
            if self.lchild is None:
                temp=self.rchild
                self=None
                return temp
            if self.rchild is None:
                temp=self.lchild
                self=None
                return temp
            node=self.rchild
            while node.lchild:
                node=node.lchild
            self.key=node.key
            self.rchild=self.rchild.deletion(node.key)
        return self
            
        
                
    def Preorder(self):
        print(self.key,end=" ")
        if self.lchild:
            self.lchild.Preorder()
        if self.rchild:
            self.rchild.Preorder()
    
    def Inorder(self):
        if self.lchild:
            self.lchild.Inorder()
        print(self.key,end=" ")
        if self.rchild:
            self.rchild.Inorder()
    
    def Postorder(self):
        if self.lchild:
            self.lchild.Inorder()
        if self.rchild:
            self.rchild.Inorder()
        print(self.key,end=" ")

Root=Node(12)
lists=[1,2,5,6,7,22,19,18,17,3,88]
for i in lists:
    Root.insert(i)
Root.Inorder()
print()
Root.Postorder()
print()
Root.Preorder()
print()
Root.Search(22)
Root.Search(12)
Root.deletion(7)
Root.Inorder()
        
        
```
